1. Write a Python Program to Display Fibonacci Sequence Using Recursion?





```python
def fib(n):
    if(n==0):
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
    

x = int(input("How many terms user wants? : "))
for i in range(x):
    print(fib(i))
```

2. Write a Python Program to Find Factorial of Number Using Recursion?

```python
def fact(n):
    if n==0:
        return 1
    else:
        return fact(n-1)*n
        
n = int(input("Enter a No. : "))
if(n>=0):
    print("Factorial of {0} is {1}".format(n,fact(n)))
else:
    print("Please enter +ve no")
```

3. Write a Python Program to calculate your Body Mass Index?

```python
m = float(input("Enter weight of person in Kg : "))
h = float(input("Enter height of person in m : "))
bmi = m/(h*h)
bmi = round(bmi,2)
print("BMI of person is : ",bmi)
if(bmi<18.5):
    print("Person is Underweight")
elif bmi >18.5 and bmi <24.9 :
    print("Normal Weight")
else:
    print("Person is Overweight")
```

5. Write a Python Program for cube sum of first n natural numbers?

```python
n = int(input("Enter the no of element, to find Sum : "))
if(n>0):
    x = int((n*(n+1))**2)/4
    print(x)
else:
    print("Enter only natural no")
```

### Alternative Method

```python
def Cube_sum(n):
    sum=0
    for i in range(1,n+1):
        sum += i**3
    return sum
n = int(input("Enter the no of element, to find Sum : "))
if(n>0):
    x = Cube_sum(n)
    print(x)
else:
    print("Enter only natural no")
```
