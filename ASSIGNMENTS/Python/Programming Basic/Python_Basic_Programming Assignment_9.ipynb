1. Write a Python program to check if the given number is a Disarium Number?






```python
def n_len(n):
    sum=0
    while(n>0):
        sum += 1
        n = n//10
    return sum

def dis_num(n):
    x=n
    rem,sum=0,0
    l = n_len(n)
    while(l):
        rem = n%10
        sum += rem**l
        n = n//10
        l -= 1
    #print(x,sum)
    if(x==sum):
        return True
    else:
        return False
    
    
n = int(input("Enter a No. : "))

if(dis_num(n)):
    print(n," is a Disarium Number")
else:
    print(n," is a Not Disarium Number")
```

2. Write a Python program to print all disarium numbers between 1 to 100?

```python
print("Disarium numbers between 1 to 100 are : \n")
for i in range(0,101):
    if(dis_num(i)):
        print(i)
```

3. Write a Python program to check if the given number is Happy Number?

```python
def sq_sum(n):
    sum=0
    while(n>0):
        x = n%10
        sum  += x*x
        n = n//10
    return sum
def happy_num(n):
    if(n==1 or n==7):
        return True
    else:
        sum,x=n,n
        while(sum>9):
            sum=sq_sum(x)
            #print(sum)
            if(sum==1 or sum==7):
                return True
            x=sum
    return False
    
```

```python
n = int(input("Enter a No. : "))
y = happy_num(n)
if(y):
    print(n," is a Happy no")
else:
    print(n," is a Not Happy no")
```

6. Write a Python program to print all pronic numbers between 1 and 100?

```python
# to verfiy any perticular no , is it pronic or not ?
def p_no(n):
    for i in range(n):
        x = i*(i+1)
        #print(x)
        if(x == n):
            return True
    else:
        return False
n = int(input("Enter a No. : "))
if(p_no(n)):
    print(n," is a pronic Number")
else:
    print(n," is a Not pronic Number")
```
