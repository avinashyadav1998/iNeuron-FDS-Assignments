1. Write a Python program to convert kilometers to miles?

```python
x = int(input("Enter distance in km : "))
y = (0.621371)*x
print("{0} km is equals to {1} in miles".format(x,y))
```

4. Write a Python program to solve quadratic equation?

```python
''''
Note - 
ax2 + bx + c
where,
a, b, and c are coefficient and real numbers and also a ≠ 0.
If a is equal to 0 that equation is not valid quadratic equation.
'''
```

#### x1 = (-b + sqrt(D))/2a
#### x2 = (-b - sqrt(D))/2a
#### D = b*b - 4ac

```python
import math

def quad(a,b,c):
    D = b**2 -4*a*c
    d = math.sqrt(abs(D))
    
    if D>0 :
        print("roots are real\n")
        x1 = (-b + d)/(2*a)
        x2 = (-b - d)/(2*a)
        print(x1, x2)
    elif D==0:
        print(" real and same roots")  
        print(-b/(2*a))
    else:
        print("Complex Roots")  
        print(- b/(2*a)," + ", d/(2*a),"i")  
        print(- b/(2*a)," - ", d/(2*a),"i")
        
        
print("a, b, and c are coefficient and real numbers, enter values ")
a = int(input("Enter a : "))
b = int(input("Enter b : "))
c = int(input("Enter c : "))

if(a<0):
    print("incorrect equation")
else:
    quad(a,b,c)
```

5. Write a Python program to swap two variables without temp variable?

```python
#using XOR
def swap_without_temp(a,b):
    a = a^b
    b = a^b
    a = a^b
    return a,b

a,b = 5,7
print("before swap : ",a,b)


a,b = 10,17
print("before swap : ",a,b)
a,b = swap_without_temp(a,b)
print("after swap : ",a,b)
```
