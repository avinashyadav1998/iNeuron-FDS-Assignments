1. Write a Python program to find words which are greater than given length k?







```python
def word_len(s,x):
    #change string to list & delimiter is space
    ans = []
    l = s.split(" ")
    print(l)
    for i in l:
        if(len(i)>x):
            #print(i)
            ans.append(i)
    #print(ans)
    return ans
    
s = input("Enter string : ")
x = int(input("Enter min length of word which you want to extract : "))
ans = word_len(s,x)
print("words which are greater than length {0} are : ".format(x),len(ans),ans)
```

2. Write a Python program for removing i-th character from a string?

```python
#method 1
def remove_char1(s,x):
    ans=""
    for i in range(len(s)):
        if(i!=x):
            ans += s[i]
    return ans
    
s = input("Enter string : ")
x = int(input("Enter position of character which you want to remove : "))
print(remove_char1(s,x))
```

```python
#method 2
def remove_char2(s,x):
    ans=""
    if(x>len(s)):
        return -1
    elif(x==len(s)-1):
        ans = s[:x]
    else:
        ans  = s[:x] + s[x+1]
    return ans
    
s = input("Enter string : ")
x = int(input("Enter position of character which you want to remove : "))
#print(remove_char2(s,x))
print(s[:x]+s[x+1:])
```

```python
#method 3
def remove_char3(s,x):
    ans = s.replace(s[x], '')
    return ans
    
s = input("Enter string : ")
x = int(input("Enter position of character which you want to remove : "))
print(remove_char3(s,x))
```

4. Write a Python to check if a given string is binary string or not?

```python
#Method 1
def check_binary1(s):
    t = {'0','1'}
    x = set(s)
    #print(t,x)
    if(t==x):
        return True
    return False

s = input("Enter string : ")
if(check_binary1(s)):
    print("string is binary")
else:
    print("string is not binary")
```

```python
#Method 2
def check_binary2 (s):
    for i in s:
        if i !="0" and i !="1":
            return False
    else:
        return True

s = input("Enter string : ")
if(check_binary2(s)):
    print("string is binary")
else:
    print("string is not binary")
```

5. Write a Python program to find uncommon words from two Strings?

```python
def uncommon_words(s1,s2):
    x1 = set(s1.split())
    x2 = set(s2.split())
    #print(x1)
    #print(x2)
    
    #apply symmetric difference
    ans = x1.symmetric_difference(x2)
    return list(ans)
    
    
s1 = input("Enter 1st string : ")
s2 = input("Enter 2nd string : ")
print("String 1 : ",s1)
print("String 2 :",s2)
print("uncommon words from above Strings",uncommon_words(s1,s2))
```

6. Write a Python to find all duplicate characters in string?

```python
def duplicate(s):
    l =[]
    d = {}
    for i in s:
        d[i] = d.get(i,0)+1
        
    for k,v in d.items():
        if(v>1):
            l.append(k)
    return l
    
    
s = input("Enter string : ")
print(s)
ans = duplicate(s)
print(" duplicate characters in string are : ",ans)
```

7. Write a Python Program to check if a string contains any special character?

```python
s_c = "[@_!#$%^&*()<>?/\|}{~:]"
s = input("Enter string : ")
print(s)
for i in s:
    if i in s_c:
        print("string contains special character")
        break
else:
    print("string doesn't contain any special character")
```
