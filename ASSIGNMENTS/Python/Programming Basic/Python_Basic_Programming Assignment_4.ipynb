1. Write a Python Program to Find the Factorial of a Number?

```python
def fact_recursive(n):
    if(n==0):
        return 1
    else:
        return fact(n-1)*n
        
n = int(input("Enter a No. : "))
if(n>=0):
    print(fact_recursive(n))
else:
    print("Please enter +ve no")
```

```python
def fact_iterative(n):
    x=1
    for i in range(1,n+1):
        x = x*i
    return x
    
n = int(input("Enter a No. : "))
if(n>=0):
    print(fact_iterative(n))
else:
    print("Please enter +ve no")
```

3. Write a Python Program to Print the Fibonacci sequence?

```python
# Program for Nth Fibonacci Term
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
    
```

```python
n = int(input("Enter nth position. : "))
if(n>=0):
    print(fib(n))
else:
    print("Please enter +ve no")
```

```python
x = int(input("How many terms user wants? : "))
for i in range(x):
    print(fib(i))
```

4. Write a Python Program to Check Armstrong Number?

```python
import math
x = int(input("Enter No. : "))
def fun(n):
    x = int(math.log10(n)+1)
    #print(x)
    
    t = n
    sum,rem=0,0
    while(n>0):
        rem = n%10
        sum += rem**x
        n = n//10
    #print(sum)
    if(sum==t):
        return True
    return False
if(fun(x)):
    print(x,"is an Armstrong number")
else:
    print(x,"is not an Armstrong number")
```

5. Write a Python Program to Find Armstrong Number in an Interval?

```python
x = int(input("Enter starting Number of Interval : "))
y = int(input("Enter Ending Number of Interval : "))
ans = []
for i in range(x,y+1):
    if(fun(i)):
        ans.append(i)

print("Armstrong Number b/w {0} - {1} are : ".format(x,y),ans)
```

6. Write a Python Program to Find the Sum of Natural Numbers?

```python
n = int(input("Enter the no of element, to find Sum : "))
if(n>0):
    x = int((n*(n+1))/2)
    print(x)
else:
    print("Enter only natural no")
```

### Alternative Method

```python
def Sq_sum(n):
    sum=0
    for i in range(1,n+1):
        #print(i,sum)
        sum += i
    return sum
n = int(input("Enter the no of element, to find Sum : "))
if(n>0):
    x = Sq_sum(n)
    print(x)
else:
    print("Enter only natural no")
```
