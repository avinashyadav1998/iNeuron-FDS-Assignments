2. Write a Python program to find the sum of all items in a dictionary?

```python
import random 
#generating dictonary
def gen_dict(s):
    marks = {}
    for i in range(5):
        marks[s+'_sub'+str(i)] = random.randint(33,100)
    return marks


#generating dictonary to use
marks = gen_dict('student1')
print(marks)
sum =0
for i in marks.values():
    sum += i
print("sum of all items in a dictionary are : ",sum)

```

3. Write a Python program to Merging two Dictionaries?

```python
#Method 1
def merge1(d1,d2):
    x  = {**d1, **d2}
    return x

def merge2(d1,d2):
    d1.update(d2)
    return d1
#generating dictonary to use
d1 = gen_dict("student1")
print(d1)
d2 = gen_dict("student2")
print(d2)

#Merging two Dictionaries1
#Method 1
print("\nMerged Dictionaries : ", merge1(d1,d2))
#Method 2
print("\nMerged Dictionaries : ", merge2(d1,d2))
```

7. Write a Python program to sort Python Dictionaries by Key or Value?

```python
orders = {
'cappuccino': 54,
'latte': 56,
'espresso': 72,
'americano': 48,
'cortado': 41
}
print(orders)
```

```python
print("Sorted by Keys : ")
sort_orders_by_key = sorted(orders.items(), key=lambda x: x[1], reverse=False)
for i in sort_orders_by_key:
    print(i[0], i[1])
```

```python
print("Sorted by Values : ")
sort_orders_by_values = sorted(orders.items(),key = lambda x:x[0])
for i in sort_orders_by_values:
    print(i[0],i[1])
```
