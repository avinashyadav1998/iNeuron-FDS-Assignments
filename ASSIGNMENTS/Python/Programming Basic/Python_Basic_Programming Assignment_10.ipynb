10. Write a Python program to Cloning or Copying a list?

```python
#1 
l1 = [x for x in l ]
print("\n : ",l1)

#2
l2 = l[:]
print("\n : ",l2)

#3
import copy
#shallow copy
l3  = copy.copy(l)
print("\n : ",l3)

#deep copy
l4  = copy.deepcopy(l)
print("\n : ",l4)
```

11. Write a Python program to Count occurrences of an element in a list?

```python
def cnt_ele1(l, x):
    cnt = 0
    for i in l:
        if (i == x):
            cnt = cnt + 1
    return cnt

def cnt_ele2(l, x):
    return l.count(x)

x = int(input("Enter element to find occurrences of it : "))
print('{} has occurred {} times'.format(x, cnt_ele1(l,x)))
print('{} has occurred {} times'.format(x, cnt_ele2(l,x)))
```
