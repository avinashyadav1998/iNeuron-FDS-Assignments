Question 1
Create a function that takes a number as an argument and returns True or False depending
on whether the number is symmetrical or not. A number is symmetrical when it is the same as
its reverse.
Examples
is_symmetrical(7227) ➞ True
is_symmetrical(12567) ➞ False
is_symmetrical(44444444) ➞ True
is_symmetrical(9939) ➞ False
is_symmetrical(1112111) ➞ True

```python
import math
def is_symmetrical(n):
    s = str(n)
    
    l = len(s)-1
    h = int(l/2)+1
    for i in range(0,h):
        #print(s[i],s[l-i])
        if(s[i] != s[l-i]):
            return False
        
    return True

n = int(input("Enter a number : "))
if(is_symmetrical(n)):
    print("number is symmetrical")
else:
    print("number is not symmetrical")
```

Question 2
Given a string of numbers separated by a comma and space, return the product of the
numbers.
Examples
multiply_nums(&quot;2, 3&quot;) ➞ 6
multiply_nums(&quot;1, 2, 3, 4&quot;) ➞ 24
multiply_nums(&quot;54, 75, 453, 0&quot;) ➞ 0
multiply_nums(&quot;10, -2&quot;) ➞ -20

```python
def multiply_nums(s):
    l = s.split(', ')
    #print(l)
    sum = 1
    for i in l:
        sum *=  int(i)
    
    return sum

s = input("Enter a string of numbers separated by a comma and space : ")
print("multiplication of nums is : ",multiply_nums(s))
```

Question 3
Create a function that squares every digit of a number.
Examples
square_digits(9119) ➞ 811181
square_digits(2483) ➞ 416649
square_digits(3212) ➞ 9414
Notes
The function receives an integer and must return an integer.

```python
def square_digits(n):
    n = str(n)
    ans = ''
    for i in n:
        x = int(i)
        ans += str(x*x)
        
    return int(ans)
    
n = int(input("Enter a no : "))
square_digits(n)
```

Question 4
Create a function that sorts a list and removes all duplicate items from it.
Examples
setify([1, 3, 3, 5, 5]) ➞ [1, 3, 5]
setify([4, 4, 4, 4]) ➞ [4]
setify([5, 7, 8, 9, 10, 15]) ➞ [5, 7, 8, 9, 10, 15]
setify([3, 3, 3, 2, 1]) ➞ [1, 2, 3]

```python
def setify(l):
    l.sort()
    print("sorted list : ",l)
    i=0
    while(i<len(l)):
        if(i==(len(l)-1)):
            break
        #print(l[i],l[i+1])
        if(l[i]==l[i+1]):
            l.remove(l[i])
            continue
            
        i += 1
        
    return l

l = [1, 3, 3, 5, 5, 6, 4,4]
print("final list : ",setify(l))
```

```python
l2 = [4, 4, 4, 4]
print("final list : ",setify(l2))
```

```python
l3 = [3, 3, 3, 2, 1,3,3,4]
print("final list : ",setify(l3))
```

Question 5
Create a function that returns the mean of all digits.
Examples
mean(42) ➞ 3
mean(12345) ➞ 3
mean(666) ➞ 6
Notes
 The mean of all digits is the sum of digits / how many digits there are (e.g. mean of digits in
512 is (5+1+2)/3(number of digits) = 8/3=2).
 The mean will always be an integer.

```python
def mean(n):
    s = str(n)
    l = len(s)
    sum = 0
    
    for i in s:
        sum += int(i)
        
    return int(sum/l)

n = int(input("Enter a number : "))
print(mean(n))
```
